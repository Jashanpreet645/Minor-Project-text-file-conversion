Compare the performance (time taken) of C, C++, Java, R, and Python programs to convert 200MB, 400MB, 600MB, 800Mb, 10000MB text files to UPPER CASE.

```python
import pandas as pd
import matplotlib.pyplot as m
import seaborn as sns
d1 = pd.read_csv("/content/Sample_Performance_Data.csv")
d1
```

```python
mark_style={
    'C (seconds)': 'o',
    'C++ (seconds)': 's',
    'Java (seconds)': '^',
    'Python (seconds)': 'D',
    'R (seconds)': 'x'
}
for lang in ['C (seconds)', 'C++ (seconds)', 'Java (seconds)','Python (seconds)', 'R (seconds)']:
  sns.lineplot(d1, x="File Size (MB)", y=lang)
  # sns.points(d1,pch = 20)
  sns.scatterplot(data=d1, x="File Size (MB)", y=lang, marker=mark_style[lang],label=lang)

m.xlabel("time taken (sec)")
m.ylabel("File Size (MB)")
m.title("Performance Comparsion")
m.legend()
m.show()
```
